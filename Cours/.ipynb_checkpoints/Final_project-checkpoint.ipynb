```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import re
from IPython.core.debugger import set_trace
import locale
from bs4 import BeautifulSoup
import requests
from sklearn import linear_model, metrics
import seaborn as sns

import ipywidgets
import geopandas
from cartopy import crs as ccrs

pd.set_option('display.max_columns', None)
```

## Chargement des classements

```python
def load_data():
    """Charge un tableau excel issu de https://www.vendeeglobe.org/fr/classement. 
    Ne doit pas être appelé directement."""
    
    df = pd.read_excel("data.xlsx", header = [2,3,4])
    date = df.columns.levels[0][0]
    locale.setlocale(locale.LC_ALL, 'fr_FR')
    date = datetime.datetime.strptime(date, "Classement du %A %d %B %Y à %Hh%M FR")
    #set_trace()
    df = df.droplevel(0, axis = 1)
    df = df.drop('Unnamed: 0_level_1', axis = 1,level = 0)
    df[('','Rank')] = df[('Rang\nRank','Unnamed: 1_level_2')]
    df[('','Voile')] = df[('Nat. / Voile\nNat. / Sail','Unnamed: 2_level_2')]
    df[('','Skipper/Bateau')] = df[('Skipper / Bateau\nSkipper / crew','Unnamed: 3_level_2')]
    df[('','Latitude')] = df[('Skipper / Bateau\nSkipper / crew','Latitude\nLatitude')]
    df[('','Longitude')] = df[('Skipper / Bateau\nSkipper / crew','Longitude\nLongitude')]
    df[('','Heure')] = date#df[('Skipper / Bateau\nSkipper / crew','Heure FR\nHour FR')]
    df[('','DTF')] = df[('DTF','Unnamed: 19_level_2')]
    df[('','DTL')] = df[('DTL','Unnamed: 20_level_2')]
    
    #On suprimme les dernières lignes 
    df = df.drop(df.index[-4:])
    
    #On supprime les premières colonnes
    df = df.drop(list(df.columns[0:6]),axis = 1)
    
    #On creer deux colonnes pour skipper et bateaux
    df[('','Skipper')] = df[('','Skipper/Bateau')].apply(lambda x: x.split(sep = '\n')[0])
    df[('','Bateau')] = df[('','Skipper/Bateau')].apply(lambda x: x.split(sep = '\n')[1])
    df[('','Voile')] = df[('','Voile')].apply(lambda x: x.split(sep = '\n')[1])
    df[('','Pays')] = df[('','Voile')].apply(lambda x: x[:3])
    df[('','Voile')] = df[('','Voile')].apply(lambda x: int(re.findall(r'\d+',x)[0]))
    #df[('','Heure')] = df[('','Heure')].apply(lambda x: pd.to_datetime([x][:-4], format='%H:%M'))
    
    df = df.drop(['Skipper/Bateau'],level = 1, axis = 1)
    df = df.drop(['DTF','DTL'],level = 0, axis = 1)
    
    #On places les 10 dernières colonnes en tête
    cols = df.columns.tolist()
    cols = cols[-10:] + cols[:-10]
    df = df[cols]
    
    return df
```

```python
# 2) Transformez ce DataFrame large en un DataFrame long
def data_wide2long():
    """Convertie les données en format long. Ne doit pas être appelé directement."""
    df = load_data()
    df = pd.melt(df,
                 id_vars=list(df.columns[:10]),
                 value_vars=list(df.columns[10:]))
    df['variable_0'] = df['variable_0'].apply(lambda x: x.split(sep = '\n')[0])
    df['variable_1'] = df['variable_1'].apply(lambda x: x.split(sep = '\n')[0])
    for i in range(10):
        #set_trace()
        df = df.rename(columns={df.columns[i]: df.columns[i][1]})
    
    return df
```

```python
def extract_data():
    """Extrait tous les tabeaux excel du site 'https://www.vendeeglobe.org/fr/classement' pour former un
    premier data set."""
    res = requests.get('https://www.vendeeglobe.org/fr/classement')
    soup = BeautifulSoup(res.content, 'html.parser')
    data = pd.DataFrame()

    for i in range(len(soup.find_all('select')[0].find_all('option')) - 1):
        try:
            url = 'https://www.vendeeglobe.org/download-race-data/vendeeglobe_'
            url = url + soup.find_all('select')[0].find_all('option')[i+1].get('value') + '.xlsx'
            print('chargement fichier de course n°:'+str(i+1) + '/' + str(len(soup.find_all('select')[0].find_all('option')) - 1))
            r = requests.get(url, allow_redirects=True)
            open('data.xlsx', 'wb').write(r.content)
        
            df = data_wide2long()
            data = data.append(df)
        except:
            print('L\'url ' + url + ' n\'a pas pu être téléchargée')
    data = data.reset_index()
    data = data.drop('index', axis = 1)
    
    data = data.dropna()
        
    return data

        #r = requests.get(url, allow_redirects=True)
```

## Extraction des fiches techniques et fusion

```python
def load_tech():
    """ Chargement des données de classement"""
    res = requests.get('https://www.vendeeglobe.org/fr/glossaire')
    soup = BeautifulSoup(res.content, 'html.parser')
    
    tech = {}

    for i in range(len(soup.find_all('div', attrs={'class': 'boats-list__popup mfp-hide'}))): # Pour chaque bateau 
        temp = {} 
        for j in range(len(soup.find_all('div', attrs={'class': 'boats-list__popup mfp-hide'})[i].find_all('li'))):
            info = soup.find_all('div', attrs={'class': 'boats-list__popup mfp-hide'})[i].find_all('li')[j].text.split(sep = ' : ')
            temp[info[0]] = info[1]
        print('chargement fichier technique n°:'+ str(i+1)+'/'+str(len(soup.find_all('div', attrs={'class': 'boats-list__popup mfp-hide'}))))
        tech[soup.find_all('span', attrs={'class': 'boats-list__skipper-name'})[i].text.title()] = temp
            
    tech = pd.DataFrame(tech).transpose().reset_index()
    tech.loc[tech['index'] == 'Arnaud Boissières', ['index']] = 'Arnaud Boissieres'
    tech.loc[tech['index'] == 'Sam Davies', ['index']] = 'Samantha Davies'
    tech.loc[tech['index'] == 'Alan  Roura', ['index']] = 'Alan Roura'
    #tech['index'] = tech['index'].astype(str).apply(lambda x: int(re.findall(r'\d+',x)[0]))
    #tech.set_index('index')
    return tech
```

```python
def merging():
    """Fusion des données techniques et des classements"""
    t = load_tech()
    l_data = extract_data()
    return pd.merge(l_data, t, left_on='Skipper', right_on='index', how='left')
```

### Corrélation et régression linéaire entre le classement (rang) et la vitesse utile (VMG) des voiliers.

Dans cette partie nous allons réaliser une régression lineaire sur les VMG calculées sur 24h afin de voir dans quelle mesure la VMG joue sur le classement.

```python
t = df.loc[(df['variable_1'] == 'VMG') & (df['variable_0'] == 'Depuis 24 heures'), ['Rank','value','variable_0']]
t = t.dropna()
x = t['Rank'].apply(lambda x: int(x)).to_numpy().reshape(-1,1)
y = t['value'].apply(lambda x: float(x.split()[0])).to_numpy().reshape(-1,1)

reg = linear_model.LinearRegression(fit_intercept=True).fit(x,y)
                     
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.set(xlabel='Rang',ylabel='VMG (kts)', title='Corrélation entre rang et VMG ' + t['variable_0'].iloc[0].lower())

#ax.scatter(df['value'],df['Rank'], marker='.', label= df['variable_0'])
ax.scatter(x,y, marker='.')#,label = df['variable_0'])
ax.scatter(x,reg.predict(x), marker='_')
#plt.ylim(0, 40)

#ax.legend(loc='upper right')
plt.show()
```

```python
#coefficiant de determination
"Le coefficiant de détermination R^2 vaut {}".format(1 - (sum((y - reg.predict(x))**2)/sum((y - np.mean(y))**2))[0])
```

Étant donnée la très grande amplitude des VMG pour une vitesse de vent la régression linéaire ne se déroule pas très bien. La distribution des vitesses à un rang donnée semble suivre une distribution gaussienne. Une piste pourrait être alors d'utiliser les processus gaussiens comme outil d'analyse. 

### Impact de la présence d'un foil sur le classement et la vitesse des voiliers.

On regarde dans cette partie l'impact d'un foil sur la vitesse. On trace dans un premier temps la vitesse des voiliers en fonction du rang avec le nombre de dérives en couleur. Puis on calcul le pourcentage de foils par rang. On fait une OLS sur ces dernières données

```python
t = df.loc[(df['variable_1'] == 'VMG') & (df['variable_0'] == 'Depuis 24 heures'),
           ['Rank','value','variable_0','Nombre de dérives']]
t = t.dropna()

t['Nombre de dérives'] = t['Nombre de dérives'].replace('foiler','foils')
t['Rank'] = t['Rank'].apply(lambda x: int(x))
t['value'] = t['value'].apply(lambda x: float(x.split()[0]))
#color = t['Nombre de dérives'] == (('foils') or ('foiler'))

#reg = linear_model.LinearRegression(fit_intercept=True).fit(x,y)
                     
fig, ax = plt.subplots(nrows=1, ncols=1, figsize = (15,10))
ax.set(xlabel='Rang',ylabel='VMG (kts)',
       title='Corrélation entre rang et VMG ' + t['variable_0'].iloc[0].lower() + ', nombre de dérive')

#ax.scatter(df['value'],df['Rank'], marker='.', label= df['variable_0'])
sns.scatterplot(data=t, x='Rank', y='value', hue='Nombre de dérives', markers = '_')

#ax.legend(loc='upper right')
plt.show()
```

```python
gp = t.groupby(by = ['Rank', 'Nombre de dérives']).count()
#['foils']
#gp['value'].apply(lambda x: )
a = []
for i in range(1,len(gp.index.levels[0])):
    a.append(100*gp['value'].loc[(i,'foils')]/gp['value'].loc[i].sum())
    
x = np.array(range(1,33)).reshape(-1,1)
    
reg = linear_model.LinearRegression(fit_intercept=True).fit(x,a)
    
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.set(xlabel='Rang',ylabel='% de skippeur avec foils', title='Corrélation entre le rang et le % de personnes avec foils')

ax.plot(x,a, marker='.',label = 'données')
ax.plot(x,reg.predict(x), marker='_',label = 'prédiction')
ax.plot(np.ones(33)*(100*t['Nombre de dérives'].value_counts()[0]/t['Nombre de dérives'].value_counts().sum()),
       label = 'valeur moyenne')

ax.legend(loc='lower left')
```

```python
#coefficiant de determination
"Le coefficiant de détermination R^2 vaut {}".format(1 - (sum((a - reg.predict(x))**2)/sum((a - np.mean(a))**2)))
```

Le coefficiant de détermination nous montre que les données ne sont pas linéaires. Bien qu'il peut sembler que la presence de foils favorise les skippers, la correlation n'est pas evidante.

### Visualisation de la distance parcourue par voilier.

Dans cette partie on regarde la distance parcourue par chaque skipper à une date donnée. La distance est en (nm)

```python
def vis_distance(mois = 11, jour = None):
    """Retourne la distance parcourue par chaque skipper à une date donnée.
    mois : int pour désigner le mois à regarder
    jour : int pour désiger le jour à regarder"""

    #recupérer les distances initiales
    a = df.loc[df['Heure'] == df['Heure'].min(),['DTF','Skipper']].drop_duplicates().reset_index(drop = True)
    a['DTF'] = a['DTF'].apply(lambda x: float(re.findall('[0-9]+.[0-9]',x)[0]))
    a = a.append(pd.DataFrame([[a['DTF'].max(), 'Charlie Dalin'], [a['DTF'].max(),'Arnaud Boissieres']], 
                              columns=['DTF','Skipper']),ignore_index = True)
    #date cherchée
    if jour == None:
        b = df.loc[df['Heure'] == df['Heure'].max(),['DTF','Skipper']].drop_duplicates().reset_index(drop = True)
    elif jour >= 15:
        b = df.loc[df['Heure'] == '2020-'+str(mois)+'-'+str(jour)+' 15:00:00',['DTF','Skipper']].drop_duplicates().reset_index(drop = True)
    else:
        b = df.loc[df['Heure'] == '2020-'+str(mois)+'-'+str(jour)+' 16:00:00',['DTF','Skipper']].drop_duplicates().reset_index(drop = True)
    b['DTF'] = b['DTF'].apply(lambda x: float(re.findall('[0-9]+.[0-9]',x)[0]))

    m = pd.merge(a, b, left_on='Skipper', right_on='Skipper', 
                 how='left').drop_duplicates().sort_values(['DTF_y']).reset_index(drop = True)
    m['Distance parcourue (nm)'] = m['DTF_x'] - m['DTF_y']
    #plt.plot(m['Skipper'],m['Distance parcourue (nm)'], rot = 60)
    m.plot(kind='scatter', x  = 'Skipper', y = 'Distance parcourue (nm)', 
           title = 'Distance parcourue pour chaque skipper',rot=75, figsize = (12,8));
```

```python
def vis_distance_mois(mois = 11):
    """Fonction pour la modélisation interactive avec ipywidgets"""
    dmin = df['Heure'].loc[df['Heure'].apply(lambda x: int(x.month)) == mois].min().day
    dmax = df['Heure'].loc[df['Heure'].apply(lambda x: int(x.month)) == mois].max().day
    ipywidgets.interact(vis_distance, mois = ipywidgets.fixed(mois), jour =(dmin,dmax,1));
```

### Distance à : 

Dans cette partie on regarde la distance entre les skippers à une date donnée. La distance est en (nm)

```python
def vis_distanceA(mois = 11, jour = None, skipper = 'Louis Burton'):
    """Retourne la distance à un skipper à une date donnée.
    mois : int pour désigner le mois à regarder
    jour : int pour désiger le jour à regarder
    skipper : str désigne le skipper"""
    
    #recupérer les distances initiales
    a = df.loc[df['Heure'] == df['Heure'].min(),['DTF','Skipper']].drop_duplicates().reset_index(drop = True)
    a['DTF'] = a['DTF'].apply(lambda x: float(re.findall('[0-9]+.[0-9]',x)[0]))
    a = a.append(pd.DataFrame([[a['DTF'].max(), 'Charlie Dalin'], [a['DTF'].max(),'Arnaud Boissieres']], 
                              columns=['DTF','Skipper']),ignore_index = True)

    #date cherchée
    if jour == None:
        b = df.loc[df['Heure'] == df['Heure'].max(),['DTF','Skipper']].drop_duplicates().reset_index(drop = True)
    elif jour >= 15:
        b = df.loc[df['Heure'] == '2020-'+str(mois)+'-'+str(jour)+' 15:00:00',['DTF','Skipper']].drop_duplicates().reset_index(drop = True)
    else:
        b = df.loc[df['Heure'] == '2020-'+str(mois)+'-'+str(jour)+' 16:00:00',['DTF','Skipper']].drop_duplicates().reset_index(drop = True)
    b['DTF'] = b['DTF'].apply(lambda x: float(re.findall('[0-9]+.[0-9]',x)[0]))

    m = pd.merge(a, b, left_on='Skipper', right_on='Skipper', 
                 how='left').drop_duplicates().sort_values(['DTF_y']).reset_index(drop = True)
    m['Distance parcourue (nm)'] = m['DTF_x'] - m['DTF_y']
    m['Distance a'] = m['Distance parcourue (nm)'] - float(m['Distance parcourue (nm)'].loc[m['Skipper'] == skipper])
    #plt.plot(m['Skipper'],m['Distance parcourue (nm)'], rot = 60)
    m.plot(kind='scatter', x  = 'Skipper', y = 'Distance a', c = list(map(lambda x: 'r' if x else 'b',list(m["Skipper"] == skipper))),
           title = 'Distance parcourue pour chaque skipper',rot=75, figsize = (12,8));
```

### Cartes avec les routes d'un ou plusieurs voiliers selon diverses projections

Dans cette partie on trace le parcours des différents skippers sur une carte du monde. Il est aussi possible de choisir une projection parmi une liste.

```python
a = df['Latitude'].apply(lambda x: float(re.findall('[0-9]+',x)[0])+(float(re.findall('[0-9]+',x)[1])/100))
b = df['Latitude'].apply(lambda x: 1 if re.findall('(S|N)',x)[0] == 'N' else -1)
df['Lat2'] = a * b

a = df['Longitude'].apply(lambda x: float(re.findall('[0-9]+',x)[0])+(float(re.findall('[0-9]+',x)[1])/100))
b = df['Longitude'].apply(lambda x: 1 if re.findall('(W|E)',x)[0] == 'E' else -1)
df['Lon2'] = a * b
```

```python
def visMap(skipper = 'Louis Burton', crs = None):
    """Fonction qui est utile pour tracer le parcours des skippers.
    skipper = str ou list ensemble des skippers étudiés
    crs : cartopy.crs. projection choisie pour la représentation"""
    world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
    if type(skipper) == list:
        df2 = df.loc[df['Skipper'].isin(skipper)]
    else:
        df2 = df.loc[df['Skipper'] == skipper]
    gdf = geopandas.GeoDataFrame(df2, geometry=geopandas.points_from_xy(df2.Lon2,df2.Lat2))
    gdf.crs = world.crs
    
    if crs != None:
        world = world.to_crs(crs.proj4_init)
        gdf = gdf.to_crs(crs.proj4_init)

    gdf.plot(ax = world.plot(color='white', edgecolor='black', figsize = (14,14)),
             markersize = 2, alpha = 1, column ='Skipper', legend = True);
    
```

### Analyse Surface de voiles

Dans cette partie, j'ai essayé de voir si la surface des voiles avait une influence sur le classement. J'ai donc essayé de voir si l'on pouvait retrouver un cluster où les meilleurs auraient une certaine surface spécifique de voile. On va pour ce faire s'interesser aux colonnes : 'Surface de voiles au près' et 'Classement en fonction de la surface des voiles'

```python
symlist = ['o', 's', 'D', 'x', '+', '*', 'p', 'v', '^']
def plot_2d(data, y=None):
    """Fonction qui plot en 2d des points avec un label
    data : numpy.array data à ploter
    y : numpy.array de même longueur que data"""

    k = np.unique(y).shape[0]
    color_blind_list = sns.color_palette("colorblind", k)
    sns.set_palette(color_blind_list)
    if y is None:
        labs = [""]
        idxbyclass = [range(data.shape[0])]
    else:
        labs = np.unique(y)
        idxbyclass = [np.where(y == labs[i])[0] for i in range(len(labs))]

    for i in range(len(labs)):
        plt.scatter(data[idxbyclass[i], 0], data[idxbyclass[i], 1],
                    c=np.reshape(color_blind_list[i], (1, -1)), s=80,
                    marker=symlist[i],label = y[i])
    plt.ylim([np.min(data[:, 1]) - 30, np.max(data[:, 1]) + 30])
    plt.xlim([np.min(data[:, 0]) - 30, np.max(data[:, 0]) + 30])
    plt.legend(loc='upper right')
    plt.xlabel('Surface de voiles au près')
    plt.ylabel('Surface de voiles au portant')
    plt.title('Classement en fonction de la surface des voiles')
    mx = np.min(data[:, 0])
    maxx = np.max(data[:, 0])
```

On regarde si les deux premier ont se retrouve quelque part dans notre espace à deux dimensions.

```python
t = df.loc[(df['variable_1'] == 'VMG') & (df['variable_0'] == 'Depuis 24 heures'), ['Rank','value','Surface de voiles au près','Surface de voiles au portant' ]]
t['Rank'] = t['Rank'].apply(lambda x: int(x))
t = t.loc[t['Rank'] < 3]

y = y = t['Rank'].to_numpy()
x1 = t['Surface de voiles au près'].apply(lambda x: int(re.findall('[0-9]+',x)[0]))
x2 = t['Surface de voiles au portant'].apply(lambda x: int(re.findall('[0-9]+',x)[0]))
x = pd.concat([x1, x2], axis=1, sort=False)
```

```python
plot_2d(x.to_numpy(),y)
```

À priori, il est difficile d'affirmer qu'une surface de voile est mieux qu'une autre. Essayons de regarder la même chose pour les dernières places.

```python
t = df.loc[(df['variable_1'] == 'VMG') & (df['variable_0'] == 'Depuis 24 heures'), ['Rank','value','Surface de voiles au près','Surface de voiles au portant' ]]
t['Rank'] = t['Rank'].apply(lambda x: int(x))
t = t.loc[t['Rank'] > 30]

y = y = t['Rank'].to_numpy()
x1 = t['Surface de voiles au près'].apply(lambda x: int(re.findall('[0-9]+',x)[0]))
x2 = t['Surface de voiles au portant'].apply(lambda x: int(re.findall('[0-9]+',x)[0]))
x = pd.concat([x1, x2], axis=1, sort=False)
```

```python
plot_2d(x.to_numpy(),y)
```

On peut voir qu'il est difficile de trouver une region de l'espace des premiers où une region de l'espace des derniers. Il semble donc a priori difficile d'affirmer qu'une certaine surface de voile est mieux qu'une autre.
